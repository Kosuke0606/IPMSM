```python
import os

from PIL import Image


def convert_images_to_grayscale(input_dir, output_dir):
    """
    指定された入力ディレクトリ内のBMP画像をすべてグレースケールに変換し、
    出力ディレクトリに保存する関数。

    Args:
        input_dir (str): 入力画像が含まれるディレクトリのパス。
        output_dir (str): グレースケール画像を保存するディレクトリのパス。
    """

    # 出力ディレクトリが存在しない場合は作成 (存在していてもエラーにならない)
    os.makedirs(output_dir, exist_ok=True)
    print(f"入力ディレクトリ: {input_dir}")
    print(f"出力ディレクトリ: {output_dir} (存在しない場合は作成されました)")

    processed_count = 0
    skipped_count = 0
    error_count = 0

    try:
        filenames = os.listdir(input_dir)
    except FileNotFoundError:
        print(f"エラー: 入力ディレクトリ '{input_dir}' が見つかりません。")
        return
    except Exception as e:
        print(
            f"エラー: 入力ディレクトリ '{input_dir}' の読み込み中にエラーが発生しました: {e}"
        )
        return

    if not filenames:
        print(f"入力ディレクトリ '{input_dir}' にはファイルがありません。")
        return

    print(
        f"\n入力ディレクトリから {len(filenames)} 個のアイテムが見つかりました。処理を開始します...\n"
    )

    for filename in filenames:
        input_path = os.path.join(input_dir, filename)

        # ファイルであり、かつ拡張子が .bmp であるかを確認 (大文字・小文字を区別しない)
        if os.path.isfile(input_path) and filename.lower().endswith(".bmp"):
            output_filename = filename  # 元のファイル名を使用
            output_path = os.path.join(output_dir, output_filename)

            try:
                # 画像を開く (withステートメントで確実に閉じる)
                with Image.open(input_path) as img:
                    # グレースケールに変換 ("L"モードは輝度)
                    gray_img = img.convert("L")
                    # 画像を保存
                    gray_img.save(output_path)
                print(f"  変換完了: {filename}  ->  {output_path}")
                processed_count += 1

            except FileNotFoundError:  # 基本的には os.path.isfile でチェック済み
                print(
                    f"  エラー: ファイルが見つかりません (予期せぬ状況): {input_path}"
                )
                error_count += 1
            except IOError as e:  # Pillowが画像を開けない、または保存できない場合
                print(
                    f"  エラー: 画像ファイル '{filename}' を処理できませんでした。詳細: {e}"
                )
                error_count += 1
            except Exception as e:
                print(
                    f"  エラー: '{filename}' の処理中に予期せぬエラーが発生しました: {e}"
                )
                error_count += 1
        else:
            if os.path.isfile(input_path):
                print(f"  スキップ (BMPファイルではありません): {filename}")
            else:
                # ディレクトリやその他の非ファイルアイテムはスキップ
                print(f"  スキップ (ディレクトリまたは非BMPファイル): {filename}")
            skipped_count += 1

    print(f"\n--- 処理完了 ---")
    print(f"  処理されたBMP画像数: {processed_count}")
    print(f"  スキップされたアイテム数: {skipped_count}")
    if error_count > 0:
        print(f"  エラーが発生した画像数: {error_count}")


if __name__ == "__main__":
    # === 設定項目 ===
    input_directory = "data/20250425_data/Material_d"
    output_directory = "data/20250425_data/Material_d_Gray"
    # ===============

    # スクリプトが置かれているディレクトリを基準にする場合 (コメントアウトされています)
    # script_dir = os.path.dirname(os.path.abspath(__file__))
    # input_directory = os.path.join(script_dir, "data/20250425_data/Material_d")
    # output_directory = os.path.join(script_dir, "data/20250425_data/Material_d_Gray")

    convert_images_to_grayscale(input_directory, output_directory)
```
